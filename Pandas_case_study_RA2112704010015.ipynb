```python
#Total Number of App Titles contain Astrology
l=len(data[data['App'].str.contains('Astrology',case = True)])
print(l)
```

```python
# Average App Rating
print(data['Rating'].mean())
```

```python
#Total Number of Unique category
print(data['Category'].nunique())
```

```python
#Category getting highest average rating
print(data.groupby('Category')['Rating'].mean().sort_values(ascending=False))
```

```python
#Find total number of Apps having 5 Star rating
star_5=len(data[data['Rating']==5.0])
print(star_5)
```

```python
# Find average Value of Reviews
data['Reviews']=data['Reviews'].replace('3.0M',3.0)
data['Reviews']=data['Reviews'].astype('float')
n_reviews=data['Reviews'].mean()
print(n_reviews)
```

```python
#Find Total Number of free and paid Apps
paid = data['Type'].value_counts()
print(paid)
```

```python
#App having max reviews
max_review_app=data[data['Reviews'].max()==data['Reviews']]['App']
print(max_review_app)
```

```python
#Diaplay Top 5 apps having Highest Reviews
index = data['Reviews'].sort_values(ascending=False).head().index
print(data.iloc[index]['App'])
```

```python
#Find Average Rating of free and paid Apps
data.groupby('Type')['Rating'].mean()
```

```python
#Display top 5 Apps Having Maximun Installs
data['Install_1'] = data['Installs'].str.replace(',','')
data['Install_1'] = data['Install_1'].str.replace('+','')
data['Install_1'] = data['Install_1'].str.replace('Free','0')
data['Install_1'] = data['Install_1'].astype('int')
index_installs = data['Install_1'].sort_values(ascending=False).head().index
print(data.iloc[index_installs]['App'])
```
